```python
import numpy as np
import pandas as pd
import os
```

```python
sample_img = os.listdir("sample")
sample_img = ["sample/" + i for i in sample_img]
sample_img
```

```python
from_caption = pd.read_csv("from_captions.csv",index_col=0)
alignment = pd.read_csv("alignment.csv",index_col=0)
```

```python
from_caption["image_id"] = from_caption["image_id"].apply(lambda x : "sample/" + x + ".png")
from_caption
```

```python
from_caption.set_index("image_id",inplace=True)
from_caption
```

```python
alignment["image_id"] = alignment["image_id"].apply(lambda x : "sample/" + x + ".png")
alignment
```

```python
alignment.set_index("image_id",inplace=True)
alignment
```

```python
sample_caption = from_caption.loc[sample_img]
sample_alignment = alignment.loc[sample_img]
```

```python
sample_caption["ans_len"] = sample_caption["answer"].apply(lambda x : len(x.split(" ")))
sample_alignment["ans_len"] = sample_alignment["answer"].apply(lambda x : len(x.split(" ")))
```

```python
sample_caption["ans_len"].plot(kind="hist")
```

```python
sample_caption["question"].to_frame().value_counts()
```

```python
# remove history questions [needless]
sample_caption = sample_caption[ sample_caption["question"].apply(lambda x : False if "history" in x else True)]
questions_with_long_ans =  sample_caption[sample_caption["ans_len"] > 80]["question"].value_counts()
questions_with_long_ans
# questions_with_small_ans = quesions_with_small_ans.index
```

```python
# i will just delete the entries if they have more than 35 words
sample_caption = sample_caption[sample_caption["ans_len"] < 60]
```

```python
sample_alignment["mode"] = "alignment"
sample_caption["mode"] = "pre_training"
```

```python

sample = pd.concat([sample_caption, sample_alignment])

# sample = sample.set_index('mode', append=True)

# sample.index.names = ['image_id', 'mode']
```

```python
# train - valid - test split
unique_patients = sample["case_id"].unique()
splits = dict.fromkeys(unique_patients , "")
for k,_ in splits.items():
    splits[k] = np.random.choice(["Train","Valid","Test"],p=[0.9,0.05,0.05])
splits
```

```python
sample["split"] = sample["case_id"].apply(lambda x : splits[x])
```
